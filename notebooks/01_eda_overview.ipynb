# NYC Airbnb Market Analysis – EDA

This notebook performs exploratory data analysis on NYC Airbnb listings to understand pricing drivers, demand patterns, and market behavior.


## 1. Problem Statement

Airbnb hosts and investors need to understand what drives pricing and demand across NYC neighborhoods.
This analysis aims to uncover patterns that can guide pricing strategy and market positioning.


## 2. Dataset

Source: Kaggle – New York City Airbnb Open Data (AB_NYC_2019)

The dataset contains information on Airbnb listings including:
- Location (borough, neighborhood, latitude, longitude)
- Room type
- Price
- Availability
- Reviews


## 3. Objectives

- Understand price distribution across NYC
- Identify key price drivers
- Compare borough and neighborhood trends
- Generate business-oriented insights
